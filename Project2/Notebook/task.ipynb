```python
employes = pd.read_csv("C:\\Users\\user\\Desktop\\Christine Mkhitaryan\\Data\\Raw\\employees_noisy.csv")
payroll = pd.read_csv("C:\\Users\\user\\Desktop\\Christine Mkhitaryan\\Data\\Raw\\payroll_noisy.csv")
```

```python
employes = employes[employes["EmployeeID"].notna() & (employes["EmployeeID"] != -1)]
employes = employes[employes["FirstName"].notna() & (employes["FirstName"] != " ") &
                    employes["LastName"].notna() & (employes["LastName"] != " ")]
employes['Department'] =employes['Department'].fillna('Unknown').str.title()
```

```python
employes["Salary"] = pd.to_numeric(employes["Salary"], errors = 'coerce') 
employes["Salary"] = employes["Salary"].where(employes["Salary"] > 0, np.nan)
employes["Salary"] = employes["Salary"].where(employes["Salary"] <= 500000, np.nan)
emp_med = employes["Salary"].where(employes["Salary"] != np.nan).median()
emp_med
employes["Salary"] = employes["Salary"].fillna(emp_med)
```

```python
employes["HireDate"].unique()
employes["HireDate"] = employes["HireDate"].replace("invalid date", np.nan)
employes["HireDate"]=employes["HireDate"].dropna()

```

```python
employes.to_csv("C:\\Users\\user\\Desktop\\Christine Mkhitaryan\\Data\\Processed\\employees_clean.csv")
```

```python
payroll = payroll[payroll["PayrollID"].notna() & (payroll["PayrollID"] != -1) &
                  payroll["EmployeeID"].notna() & (payroll["EmployeeID"] != -1)]
payroll["HoursWorked"] = pd.to_numeric(payroll["HoursWorked"], errors = 'coerce') 
payroll["HoursWorked"] = payroll["HoursWorked"].where(payroll["HoursWorked"] > 0, np.nan)
payroll["HoursWorked"].unique()
payroll_mean = payroll["HoursWorked"].where(payroll["HoursWorked"] != np.nan).mean()
payroll["HoursWorked"]= payroll["HoursWorked"].fillna(payroll_mean)
payroll["HoursWorked"].unique()

```

```python
payroll["Bonus"] = pd.to_numeric(payroll["Bonus"], errors = 'coerce') 
payroll["Bonus"] = payroll["Bonus"].where(payroll["Bonus"] > 0, 0)
payroll["Bonus"].unique()
payroll["Bonus"]= payroll["Bonus"].fillna(0)
payroll["Bonus"].unique()
```

```python
payroll["Deductions"] = pd.to_numeric(payroll["Deductions"], errors = 'coerce') 
payroll["Deductions"] = payroll["Deductions"].where(payroll["Deductions"] > 0, np.nan)
payroll_bonus_mean = payroll["Deductions"].where(payroll["Deductions"] != np.nan).median()
payroll["Deductions"]= payroll["Deductions"].fillna(payroll_bonus_mean)
payroll["Bonus"].unique()
```

```python

payroll_final = payroll.merge(employes[['EmployeeID', 'Salary']], on='EmployeeID', how='inner')
payroll_final.loc[payroll_final['Deductions'] > payroll_final['Salary'], 'Deductions'] = payroll_final['Salary']
payroll_validated = payroll_final.drop(columns=['Salary'])
payroll_validated.to_csv('C:\\Users\\user\\Desktop\\Christine Mkhitaryan\\Data\\Processed\\payroll_validated.csv', index=False)
```

```python
conn = sqlite3.connect("C:\\Users\\user\\Desktop\\Christine Mkhitaryan\\Data\\DB\\EmployeeManagement.db")
curs = conn.cursor()
```

```python
                   
curs.execute("DROP TABLE IF EXISTS Payroll")
curs.execute("DROP TABLE IF EXISTS SalaryHistory")
curs.execute("DROP TABLE IF EXISTS Employees")

curs.execute('''
                   CREATE TABLE Employees (
    EmployeeID INTEGER PRIMARY KEY NOT NULL,
    FirstName TEXT NOT NULL, 
    LastName TEXT NOT NULL,
    Department TEXT NOT NULL, 
    HireDate TEXT NOT NULL,
    Salary REAL NOT NULL CHECK (Salary > 0 AND Salary <= 500000))
                   ''')

curs.execute('''
             CREATE TABLE Payroll (
    PayrollID INTEGER PRIMARY KEY NOT NULL,
    EmployeeID INTEGER NOT NULL, 
    PayPeriod TEXT NOT NULL,
    HoursWorked REAL NOT NULL CHECK (HoursWorked >= 0 AND HoursWorked <= 168),
    Bonus REAL NOT NULL CHECK (Bonus >= 0),
    Deductions REAL NOT NULL CHECK (Deductions >= 0),
    FOREIGN KEY (EmployeeID) REFERENCES Employees(EmployeeID))
             ''')

curs.execute( '''
        CREATE TABLE SalaryHistory (
        HistoryID  INTEGER PRIMARY KEY AUTOINCREMENT,
        EmployeeID INTEGER NOT NULL,
        OldSalary REAL,
        NewSalary REAL NOT NULL,
        ChangeDate TEXT DEFAULT CURRENT_TIMESTAMP NOT NULL,
        FOREIGN KEY (EmployeeID) REFERENCES Employees(EmployeeID))
                  ''' )  
                   
                   
```

```python
curs.execute(""
"DROP TRIGGER IF EXISTS LogSalaryChange;")

curs.execute("""
CREATE TRIGGER LogSalaryChange
AFTER UPDATE OF Salary ON Employees
FOR EACH ROW
WHEN OLD.Salary IS NOT NEW.Salary AND NEW.Salary IS NOT NULL
BEGIN
    INSERT INTO SalaryHistory (EmployeeID, OldSalary, NewSalary, ChangeDate)
    VALUES (OLD.EmployeeID, OLD.Salary, NEW.Salary, CURRENT_TIMESTAMP);
END;
""")
```

```python
curs.execute("DROP VIEW IF EXISTS DepartmentPayrollSummary;")

curs.execute("""
CREATE VIEW DepartmentPayrollSummary AS
SELECT 
    e.Department,
    COUNT(DISTINCT e.EmployeeID) AS TotalEmployees,
    SUM(p.HoursWorked) AS TotalHoursWorked,
    SUM(p.Bonus) AS TotalBonus,
    AVG(DISTINCT e.Salary) AS AverageSalary
FROM Employees e
LEFT JOIN Payroll p ON e.EmployeeID = p.EmployeeID
GROUP BY e.Department;
""")
```
