```python
customers = customers[customers["CustomerID"].notna() & (customers["CustomerID"] != -1)]
customers=customers[customers["FirstName"].notna() & (customers["FirstName"] != "" ) &
                    customers["LastName"].notna() & (customers["LastName"] != "")]
customers["Email"] = customers["Email"].str.lower()
customers = customers[customers["Email"].notna() & (customers["Email"] != "")]
```

```python
customers.to_csv(f'C:\\Users\\user\\Desktop\\Exam\\Data\\Processed\\customers_clean.csv', index=False)
```

```python
orders = orders[orders["OrderID"].notna() & (orders["OrderID"] != -1) & 
                orders["CustomerID"].notna() & (orders["CustomerID"] != -1)]
orders["OrderAmount"]= pd.to_numeric(orders["OrderAmount"], errors='coerce')
orders.loc[orders["OrderAmount"] < 0, "OrderAmount"] = pd.NA
valid_status=["Pending","Shipped", "Cancelled"]
orders["Status"]=orders["Status"].where(orders["Status"].isin(valid_status), "Pending")
orders
```
