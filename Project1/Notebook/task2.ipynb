```python
import pandas as pd
import sqlite3
import os
```

```python
customers = pd.read_csv("C:\\Users\\user\\Desktop\\Exam\\Data\\Row\\customers_noisy.csv")
customers
```

```python
customers = customers[
    customers["CustomerID"].notna() & (customers["CustomerID"] != -1)
]
customers = customers[
    customers["FirstName"].notna() | customers["LastName"].notna()
]
customers["Email"] = customers["Email"].str.lower()
customers = customers[customers["Email"].notna()]
customers
```

```python
customers.to_csv(f"C:\\Users\\user\\Desktop\\Exam\\Data\\Processed\\customers_clean.csv", index=False)
```

```python
orders = pd.read_csv("C:\\Users\\user\\Desktop\\Exam\\Data\\Row\\orders_noisy.csv")
orders
```

```python
orders = orders[
    orders["OrderID"].notna() & (orders["OrderID"] != -1) &
    orders["CustomerID"].notna() & (orders["CustomerID"] != -1)
]
orders["OrderAmount"] = pd.to_numeric(orders["OrderAmount"], errors="coerce")
orders.loc[orders["OrderAmount"] < 0, "OrderAmount"] = pd.NA
orders["OrderAmount"].fillna(orders["OrderAmount"].median(), inplace=True)
orders
```

```python
allowed_status = ["Pending", "Shipped", "Cancelled"]
orders["Status"] = orders["Status"].where(orders["Status"].isin(allowed_status), "Pending")
orders.to_csv(f"C:\\Users\\user\\Desktop\\Exam\\Data\\Processed\\orders_clean.csv", index=False)
orders
```

```python
conn = sqlite3.connect("C:\\Users\\user\\Desktop\\Exam\\Data\\DB\\CustomerOrders.db")
cursor = conn.cursor()
```

```python
customers_clean.to_sql("Customers", conn, if_exists="append", index=False)
orders_validated.to_sql("Orders", conn, if_exists="append", index=False)
```

```python
cursor.executescript("""
DROP TABLE IF EXISTS OrderStatusLog;
DROP TABLE IF EXISTS Orders;
DROP TABLE IF EXISTS Customers;
""")

cursor.executescript("""
CREATE TABLE Customers (
    CustomerID INTEGER PRIMARY KEY,
    FirstName TEXT,
    LastName TEXT,
    Email TEXT UNIQUE,
    JoinDate TEXT
);

CREATE TABLE Orders (
    OrderID INTEGER PRIMARY KEY,
    CustomerID INTEGER,
    OrderDate TEXT,
    OrderAmount REAL,
    Status TEXT,
    FOREIGN KEY (CustomerID) REFERENCES Customers(CustomerID)
);

CREATE TABLE OrderStatusLog (
    LogID INTEGER PRIMARY KEY AUTOINCREMENT,
    OrderID INTEGER,
    OldStatus TEXT,
    NewStatus TEXT,
    ChangeDate TEXT DEFAULT CURRENT_TIMESTAMP
);
""")
```

```python
cursor.executescript("""
CREATE TRIGGER LogOrderStatusChange
AFTER UPDATE OF Status ON Orders
FOR EACH ROW
WHEN OLD.Status <> NEW.Status
BEGIN
    INSERT INTO OrderStatusLog (
        OrderID,
        OldStatus,
        NewStatus,
        ChangeDate
    )
    VALUES (
        OLD.OrderID,
        OLD.Status,
        NEW.Status,
        CURRENT_TIMESTAMP
    );
END;
               """)
```

```python
cursor.executescript("""
CREATE VIEW CustomerRevenueSummary AS
SELECT
    c.CustomerID,
    c.FirstName || ' ' || c.LastName AS FullName,
    COUNT(o.OrderID) AS TotalOrders,
    SUM(
        CASE
            WHEN o.Status = 'Shipped'
            THEN o.OrderAmount
            ELSE 0
        END
    ) AS TotalRevenue
FROM Customers c
LEFT JOIN Orders o
    ON c.CustomerID = o.CustomerID
GROUP BY c.CustomerID;
                     """)
```
